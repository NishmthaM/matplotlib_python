```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[2,3,2,4,5]

plt.plot(x,y,color="b",alpha=0.4,label="Line 1")
plt.savefig("ml.png",dpi=2000,facecolor="g")
plt.title("Simple Line Plot")
plt.xlabel("X-axis",fontsize=15)
plt.ylabel("Y-axis",fontsize=15)
plt.legend()
plt.show()
```
