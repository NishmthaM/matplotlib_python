**Stack plot or Staked area plot**

**multiple quantities change over time while also showing their combined total.**

**Instead of drawing separate lines for each dataset, the values are stacked on top of each other, forming colored areas.**

```python
# Stack plot represent the area

import matplotlib.pyplot as plt

x=[10,12,13,14,15]
area=[16,15,14,17,18]
area1=[10,20,30,40,50]
area2=[11,12,13,14,15]

plt.stackplot(x,area,area1,area2)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,12,13,14,15]
area1=[16,15,14,17,18]
area2=[10,20,30,40,50]
area3=[11,12,13,14,15]

lst=["area1","area2","area3"]

plt.stackplot(x,area1,area2,area3,labels=lst,colors=["purple","pink","blue"])
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,12,13,14,15]
area1=[16,15,14,17,18]
area2=[10,20,30,40,50]
area3=[11,12,13,14,15]

lst=["area1","area2","area3"]

plt.stackplot(x,area1,area2,area3,labels=lst,colors=["purple","pink","blue"],baseline="zero")
plt.legend()
plt.show()
```

**Step plot**

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=[11,12,13,14]

plt.step(x,y,color="b",marker="o",ms=10,mfc="g") # ms=marker size ,mfc=marker face color

plt.title("Step Plot")
plt.xlabel("x-axis",fontsize=15)
plt.ylabel("y-axis",fontsize=15)
plt.grid()
plt.show()
```

**Fill_Between plot**

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[6,7,8,9,10]

plt.plot(x,y,color="b")

plt.fill_between(x,y)

plt.title("Fill_between")
plt.xlabel("x-axis",fontsize=15)
plt.ylabel("y-axis",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[6,7,8,9,10]

plt.plot(x,y,color="b")

plt.fill_between(x=[2,4],y1=6,y2=8,color="purple",alpha=0.4)

plt.title("Fill_between")
plt.xlabel("x-axis",fontsize=15)
plt.ylabel("y-axis",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4,5])
y=np.array([6,7,8,9,10])

plt.plot(x,y,color="b")

plt.fill_between(x,y,where=(x>=2) & (x<=4),color="b",alpha=0.3)

plt.title("Fill_between")
plt.xlabel("x-axis",fontsize=15)
plt.ylabel("y-axis",fontsize=15)
plt.show()
```
