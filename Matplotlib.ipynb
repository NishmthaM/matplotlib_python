```python
import matplotlib.pyplot as plt

x=["Python","Java","Dart","C","C++"]
y=[90,88,85,80,82]

plt.xlabel("Language",fontsize=10)
plt.ylabel("Popularity percent")
plt.title("Bar Plot")

plt.bar(x,y,width=0.5,color="violet",align="edge",edgecolor="blue",linewidth=3)
plt.show()
```

```python
plt.bar(x,y,width=0.5,color="violet",align="edge",edgecolor="blue",linewidth=3,linestyle=":",alpha=0.5)  # dotted linestyle  // alpha value 0-1
plt.show()
```

```python
plt.bar(x,y,width=0.5,color="violet",align="edge",edgecolor="blue",linewidth=3,alpha=0.5,label="programming lanaguages")
plt.legend() # for the lable parameter
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=["Python","Java","Dart","C","C++"]
y=[90,88,85,80,82]
z=[80,62,85,45,69]
width=0.2

p=np.arange(len(x))
p1=[j+width for j in p]


plt.xlabel("Language",fontsize=10)
plt.ylabel("Popularity percent")
plt.title("Bar Plot")

plt.bar(p,y,width,color="violet")
plt.bar(p1,z,width,color="brown")  # avoid overlapping

plt.xticks(p+width/2,x,rotation=45)  # width/2 is to get names in the middle of the 2 bars
plt.show()
```

```python
import matplotlib.pyplot as plt

x=["Python","Java","Dart","C","C++"]
y=[90,88,85,80,82]

plt.xlabel("Language",fontsize=10)
plt.ylabel("Popularity percent")
plt.title("Bar Plot")

plt.barh(x,y,color="violet",align="edge")
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[2,4,6,8,10]

color=["r","b","g","y","b"]
size=[300,250,400,420,260]

plt.scatter(x,y,c=color,s=size,alpha=0.5,marker="*",edgecolor="purple",linewidth=2)

plt.title("Scatter plot",fontsize=15)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[2,4,6,8,10]

color=[0,12,24,38,24] # can give the color by numbeers range 1 - 100
size=[300,250,400,420,260]
plt.scatter(x,y,c=color,s=size,cmap="BrBG") # many color bar are there

t=plt.colorbar()
t.set_label("Color Bar")

plt.title("Scatter plot",fontsize=15)
plt.show()
```
